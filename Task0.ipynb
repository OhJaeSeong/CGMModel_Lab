```python
import pandas as pd

origin_data1 = pd.read_csv('data/DA0927_STP1.csv', encoding='cp949')
origin_data2 = pd.read_csv('data/DA0927_STP2.csv', encoding='cp949')
origin_data2
```

#### 

```python
columns_1 = ["Meas?", "G(mg/dL)", "LD1 Temp", "LD123_Bias_Av", "mPD1_dS", "T-rPD_L1dS", "R-rPD_L1dS"]
columns_2 = ["Meas?", "G(mg/dL)", "LD2 Temp", "LD123_Bias_Av", "mPD2_dS", "T-rPD_L2dS", "R-rPD_L2dS"]
columns_3 = ["Meas?", "G(mg/dL)", "LD3 Temp", "LD123_Bias_Av", "mPD3_dS", "T-rPD_L3dS", "R-rPD_L3dS"]

data_S1 = origin_data1[columns_1]
data_S1 = pd.concat([data_S1, origin_data2[columns_1]])

data_S2 = origin_data1[columns_2]
data_S2 = pd.concat([data_S2, origin_data2[columns_2]])

data_S3 = origin_data1[columns_3]
data_S3 = pd.concat([data_S3, origin_data2[columns_3]])

data_S1
```

```python
new_columns = ["Mean", "G(mg/dL)", "LD Temp", "LD123_Bias_Av", "mPD_dS", "T-rPD_dS", "R-rPD_dS"]

data_S1.columns = new_columns
data_S2.columns = new_columns
data_S3.columns = new_columns

# 결측치 제거
data_S1 = data_S1[data_S1["Mean"] != False]
data_S1 = data_S1[data_S1["mPD_dS"] != 0]
data_S1 = data_S1[data_S1["G(mg/dL)"] >= 0]

data_S2 = data_S2[data_S2["Mean"] != False]
data_S2 = data_S2[data_S2["mPD_dS"] != 0]
data_S2 = data_S2[data_S2["G(mg/dL)"] >= 0]

data_S3 = data_S3[data_S3["Mean"] != False]
data_S3 = data_S3[data_S3["mPD_dS"] != 0]
data_S3 = data_S3[data_S3["G(mg/dL)"] >= 0]

data_S1 = data_S1.drop(labels='Mean',axis=1)
data_S2 = data_S2.drop(labels='Mean',axis=1)
data_S3 = data_S3.drop(labels='Mean',axis=1)

data_S1.to_csv('data/data_S1.csv', index=False, encoding='utf-8')
data_S2.to_csv('data/data_S2.csv', index=False, encoding='utf-8')
data_S3.to_csv('data/data_S3.csv', index=False, encoding='utf-8')

data_S1
```
