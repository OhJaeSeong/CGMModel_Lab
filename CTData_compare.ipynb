#### 실제 데이터 2개 유사도 비교

```python
import pandas as pd

origin_data1 = pd.read_csv('data/CT1st_Case1.csv', encoding='cp949')
origin_data2 = pd.read_csv('data/CT1st_Case2.csv', encoding='cp949')

columns = ["LD1 Temp", "LD2 Temp", "LD3 Temp", "Rx1 Temp", "Rx2 Temp", "S1 T(C)", "S2 T(C)", "S3 T(C)", "FR Mon"]
data1 = origin_data1[columns]
data2 = origin_data2[columns]
```

```python
for col in columns:
    count = 0
    for i in range(0, len(data1)):
        if data1[col][i] == data2[col][i]:
            continue
        else:
            count += 1
    print(col, " : ", count)
```

```python
sensor_column2 = ["LD_Bias_Av", "mPDdSr31", "mPDdSr32", "mPDdSr21", "T-rPDdSr31", "T-rPDdSr32", "T-rPDdSr21"]
data_sam2 = origin_data2[sensor_column2]
data_sel2 = data_sam2[data_sam2["LD_Bias_Av"] > 0]
data_sel2 = data_sel2[data_sel2["mPDdSr31"] > 0]
data_sel2 = data_sel2[data_sel2["mPDdSr32"] > 0]
data_sel2 = data_sel2[data_sel2["mPDdSr21"] > 0]

sensor_column1= ["LD_Bias_Av", "mPD1_dS", "mPD2_dS", "mPD3_dS", "T-rPD_L1dS", "T-rPD_L2dS", "T-rPD_L3dS"]
data_sam1 = origin_data1[sensor_column1]
data_sel1 = data_sam1[data_sam1["LD_Bias_Av"] > 0]
data_sel1 = data_sel1[data_sel1["mPD1_dS"] > 0]
data_sel1 = data_sel1[data_sel1["mPD2_dS"] > 0]
data_sel1 = data_sel1[data_sel1["mPD3_dS"] > 0]

data_sel2.columns = data_sel1.columns

print(len(data_sel2))
print(len(data_sel1))
```

```python
data_sel1 = data_sel1.reset_index(drop=True)
data_sel2 = data_sel2.reset_index(drop=True)

for col in sensor_column1:
    count = 0
    for i in range(0, len(data_sel1)):
        if data_sel1[col][i] == data_sel2[col][i]:
            continue
        else:
            count += 1
    
    print(col, " : ", count)
```
